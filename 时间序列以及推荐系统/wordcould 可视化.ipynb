```python
import pandas as pd
from wordcloud import WordCloud
from matplotlib import pyplot as plt
from nltk.tokenize import word_tokenize
```

```python
#genreate_from_frequencies
data=pd.read_csv('C:/Users/zhouwei/Desktop/L2/MarketBasket/Market_Basket_Optimisation.csv',header=None)
tranc={}
total=[]
for i in range(data.shape[0]):
    tem=[]
    for j in range(data.shape[1]):
        item=str(data.values[i][j])
        if item !='nan':
            tem.append(item)
            if item not in tranc:
                tranc[item]=1
            else:
                tranc[item]+=1
    total.append(tem)
word=WordCloud(background_color='black',max_words=100)
word.generate_from_frequencies({w:c for w,c in tranc.items()})
plt.imshow(word)
print(sorted(tranc.items(),key=lambda x:x[0],reverse=True)[:10])
```

```python
#generate
tranc={}
total=[]
for i in range(data.shape[0]):
    tem=[]
    for j in range(data.shape[1]):
        item=str(data.values[i][j])
        if item !='nan':
            tem.append(item)
            if item not in tranc:
                tranc[item]=1
            else:
                tranc[item]+=1
    total.append(tem)
word2=WordCloud(background_color='black',max_words=100)
toc=[]
for item in total:
    toc.append(" ".join(item))
to="".join(toc)
cutword=word_tokenize(to)
cutted_word=" ".join(cutword)
word2.generate(cutted_word)
plt.imshow(word)
```
