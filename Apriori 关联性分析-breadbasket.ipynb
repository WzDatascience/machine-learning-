```python
from mlxtend.frequent_patterns import apriori
from mlxtend.frequent_patterns import association_rules
df=pd.read_csv('C:/Users/zhouwei/Desktop/code/BreadBasket/BreadBasket_DMS.csv')
```

```python
df.Item=df.Item.str.lower()
```

```python
df=df.drop(df[df.Item=='none'].index)
hot_encoded_df=df.groupby(['Transaction','Item'])['Item'].count().unstack().reset_index().fillna(0).set_index('Transaction')

hot_encoded_df
```

```python
frequent_itemsets = apriori(hot_encoded_df, min_support=0.02, use_colnames=True)
```

```python
frequent_itemsets
```

```python
rules=association_rules(frequent_itemsets,min_threshold=0.5,metric='lift')
```
